Random Forest Classifiers using Pyspark + MLlib
===============================================

Introduction
------------

This notebook demonstrates a random forest classifier using Pyspark and MLlib.

Spark is a cluster computing framework that offers a significant performance improvement over Hadoop and includes machine learning, streaming, and graph operation capabilities. Spark is the ideal framework for interactive data exploration, since it is capable of parallelizing advanced analytics, it offers near real-time interactivity, and it comes in many flavors that include Python, Java, Scala and R. 

Throghout the last couple of years we have observed an interesting convergence of big-data and machine learning, and I believe that Spark is the ideal platform to bridge both worlds together. Traditional learning for large amounts of data usually involves batch processing of files (eg. text, images, or videos), by breaking them apart into smaller blocks and distributing them across clusters, followed by running tranining and prediction locally with frameworks like scikit-learn or caffee. The problem with this approach is that the classifiers run locally, and the computing capabilities are limited by RAM and CPU/GPU performance. Spark in the other hand is not limited to data distribution but is also capable of parallelizing learning tasks across clusters (eg. feature generation, learning, and prediction). Spark seems to be ideal for ensemble learning and graph based operations and one of the questions that is open to debate is how can Spark improve deep learning architectures with layers that are fully connected.

Instructions to Install Spark
-----------------------------

To install PySpark locally, you need to download the latest version from [the official Spark website](http://spark.apache.org/downloads.html) and you have to select the package type: `Prebuilt with user-provided Hadoop`. Once you download and untar Spark, you have to set the following system variables in `.bashrc`:

    >>export HADOOP_CONF_DIR = <location-of-hadoop-conf-files>
    >>export HADOOP_PATH = <locationn-of-hadoop>
    >>export PATH=$PATH:HADOOP_PATH/bin:$HADOOP_PATH/sbin

Verify that the hadoop executables are in your PATH by typing:

    >>hadoop version

Then, add the following system variable to `.bashrc`:

    >>export SPARK_DIST_CLASSPATH=$(hadoop classpath)

Go to the `bin` folder in your Spark location, execute `./pyspark`, and make sure that it runs without problems. That's it! Now you're ready to configure IPython.

NOTE: I am running Spark locally for testing purposes, but ideally you would deploy Spark to Amazon EC, Google Cloud, Azure, or any other cloud computing platform.


Instructions to Configure IPython
---------------------------------

Pyspark combined with IPython notebooks is the ideal match for interactive data analysis. To configure IPython, follow the next steps [1]:

Create an IPython profile for Pyspark:

    >>ipython profile create pyspark

Set the following environment variable in `.bashrc`:

    >>export SPARK_HOME=<location-of-spark>

Create the file `~/.ipython/profile_pyspark/startup/00-pyspark-setup.py` containing the following code:

    spark_home = os.environ.get('SPARK_HOME', None)
    sys.path.insert(0, spark_home + "/python")
    
    # Add the py4j to the path.
    # You may need to change the version number to match your install
    sys.path.insert(0, os.path.join(spark_home, 'python/lib/py4j-0.8.2.1-src.zip'))
    
    # Initialize PySpark to predefine the SparkContext variable 'sc'
    execfile(os.path.join(spark_home, 'python/pyspark/shell.py'))

Start your IPython notebook with the pyspark profile:

    >>ipython notebook --profile=pyspark

And that's it, you're set!
