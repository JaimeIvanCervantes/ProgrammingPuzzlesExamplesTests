```python
%matplotlib inline
import os, struct
from array import array as pyarray
from numpy import append, array, int8, uint8, zeros
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import cv2

# Load NIST (http://g.sweyla.com/blog/2012/mnist-numpy/)
def load_mnist(dataset="training", digits=np.arange(10), path="."):
    """
    Loads MNIST files into 3D numpy arrays

    Adapted from: http://abel.ee.ucla.edu/cvxopt/_downloads/mnist.py
    """

    if dataset == "training":
        fname_img = os.path.join(path, 'train-images-idx3-ubyte')
        fname_lbl = os.path.join(path, 'train-labels-idx1-ubyte')
    elif dataset == "testing":
        fname_img = os.path.join(path, 't10k-images-idx3-ubyte')
        fname_lbl = os.path.join(path, 't10k-labels-idx1-ubyte')
    else:
        raise ValueError("dataset must be 'testing' or 'training'")

    flbl = open('train-labels-idx1-ubyte\\train-labels.idx1-ubyte','rb')#fname_lbl, 'rb')
    magic_nr, size = struct.unpack(">II", flbl.read(8))
    lbl = pyarray("b", flbl.read())
    flbl.close()

    fimg = open('train-images-idx3-ubyte\\train-images.idx3-ubyte','rb')#fname_img, 'rb')
    magic_nr, size, rows, cols = struct.unpack(">IIII", fimg.read(16))
    img = pyarray("B", fimg.read())
    fimg.close()

    ind = [ k for k in range(size) if lbl[k] in digits ]
    N = len(ind)

    images = zeros((N, rows, cols), dtype=uint8)
    labels = zeros((N, 1), dtype=int8)
    for i in range(len(ind)):
        images[i] = array(img[ ind[i]*rows*cols : (ind[i]+1)*rows*cols ]).reshape((rows, cols))
        labels[i] = lbl[ind[i]]

    return images, labels


a = np.matrix([[1,3],[2,4]])
print np.matrix.sum(a)

images, labels = load_mnist('training', digits=[2])

#mean_img = images.mean(axis=0)
#mean_img = images[5].astype(np.float32)
mean_img = images[0]

print images.shape
print "Min: {}".format(mean_img.min())
print "Max: {}".format(mean_img.max())

print mean_img.dtype
#mean_img /= 255.0

div_img = abs(mean_img[0:14, 0:14]-mean_img[14:28, 14:28])
print "Sym: {}".format(np.sum(div_img))

ret,thresh = cv2.threshold(mean_img,127,255,0)
contours, hierarchy = cv2.findContours(thresh, 1, 2)
zero_img = np.zeros([28,28])
#cv2.drawContours(zero_img, contours[1], -1, (255,255,255), 1)
M = cv2.moments(contours[0])

cx = int(M['m10']/M['m00'])
cy = int(M['m01']/M['m00'])

print "cx={}".format(cx)
print "cy={}".format(cy)

# Convex hull defects
hull = cv2.convexHull(contours[1],returnPoints = False)
defects = cv2.convexityDefects(contours[1],hull)

for i in range(defects.shape[0]):
    s,e,f,d = defects[i,0]
    start = tuple(contours[1][s][0])
    end = tuple(contours[1][e][0])
    far = tuple(contours[1][f][0])
    #cv2.line(zero_img,start,end,[255,255,255],1)
    #cv2.circle(zero_img,far,1,[255,255,255],-1)
    
#plt.imshow(div_img, cmap=plt.cm.gray)
plt.imshow(mean_img, cmap=plt.cm.gray)
plt.Circle((cx,cy),2,color='r')
plt.show()

```

```python
# Calculate features
labs = []

images, labels = load_mnist('training', digits=[2])

cxs = []
cys = []
for i in range(500):
    img = images[i]
    labs.append(labels[i])
    
    ret,thresh = cv2.threshold(img,127,255,0)
    contours, hierarchy = cv2.findContours(thresh, 1, 2)
    zero_img = np.zeros([28,28])
    M = cv2.moments(contours[0])

    if M['m00'] > 0:
        cxs.append(int(M['m10']/M['m00']))
        cys.append(int(M['m01']/M['m00']))
    else: 
        cxs.append(0)
        cys.append(0)

images, labels = load_mnist('training', digits=[0])
        
for i in range(500):
    img = images[i]
    labs.append(labels[i])
    
    ret,thresh = cv2.threshold(img,127,255,0)
    contours, hierarchy = cv2.findContours(thresh, 1, 2)
    zero_img = np.zeros([28,28])
    M = cv2.moments(contours[0])

    if M['m00'] > 0:
        cxs.append(int(M['m10']/M['m00']))
        cys.append(int(M['m01']/M['m00']))
    else: 
        cxs.append(0)
        cys.append(0)

print len(labs)
print len(cxs)
print len(cys)

'''
for i in range(len(labs)):
    print "label: {}, cx: {}, cy: {}".format(labs[i], cxs[i], cys[i])
'''   
```

```python
# Training
#x_i = [1.0,2.0]
w_ij = [[0.3,0.3], [0.3,0.3], [0.3,0.3], [0.3,0.3]]
w_jk = [0.8,0.8,0.8,0.8]
eta = 0.2
bias = 0.0
#des = 0.7
y_j = [0.0,0.0,0.0,0.0]
delta_j = [0.0,0.0]

w_t_ij = w_ij
w_t_jk = w_jk
epochs = 100

for s in range(len(labs)):
    x_i = [cxs[s],cys[s]]
    des = float(labs[s])

    for t in range(epochs):
        # Feed forward
        for j in range(len(w_t_ij)):
            A=np.dot(w_t_ij[j],x_i)+bias
            y_j[j]=1.0/(1.0+np.exp(-A))

        x_j=y_j
        A=np.dot(w_t_jk,x_j)+bias
        y_k=1.0/(1.0+np.exp(-A))
        e=des-y_k
        delta_k = e*(1.0-y_k)*y_k
        w_t_jk=w_t_jk+eta*delta_k*np.array(x_j)

        # Back propagation
        for j in range(len(w_t_ij)):
            delta_j = (1-y_j[j])*y_j[j]*(delta_k*w_jk[j])
            w_t_ij[j] =w_t_ij[j]+eta*delta_j*np.array(x_i)
            
```
