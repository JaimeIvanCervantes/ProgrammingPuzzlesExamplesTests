```python
# Training
#x_i = [1.0,2.0]
w_ij = [[0.3,0.3], [0.3,0.3], [0.3,0.3], [0.3,0.3]]
w_jk = [0.8,0.8,0.8,0.8]
eta = 0.2
bias = 0.0
#des = 0.7
y_j = [0.0,0.0,0.0,0.0]
delta_j = [0.0,0.0]

w_t_ij = w_ij
w_t_jk = w_jk
epochs = 100

for s in range(len(labs)):
    x_i = [cxs[s],cys[s]]
    des = float(labs[s])

    for t in range(epochs):
        # Feed forward
        for j in range(len(w_t_ij)):
            A=np.dot(w_t_ij[j],x_i)+bias
            y_j[j]=1.0/(1.0+np.exp(-A))

        x_j=y_j
        A=np.dot(w_t_jk,x_j)+bias
        y_k=1.0/(1.0+np.exp(-A))
        e=des-y_k
        delta_k = e*(1.0-y_k)*y_k
        w_t_jk=w_t_jk+eta*delta_k*np.array(x_j)

        # Back propagation
        for j in range(len(w_t_ij)):
            delta_j = (1-y_j[j])*y_j[j]*(delta_k*w_jk[j])
            w_t_ij[j] =w_t_ij[j]+eta*delta_j*np.array(x_i)
            
```

```python
# Generate three random clusters of 2D data
N_c = 200
A = 0.6*np.random.randn(N_c, 2)+[1, 1]
B = 0.6*np.random.randn(N_c, 2)+[3, 3]
C = 0.6*np.random.randn(N_c, 2)+[3, 0]
X = np.hstack((np.ones(3*N_c).reshape(3*N_c, 1), np.vstack((A, B, C))))
Y = np.vstack(((np.zeros(N_c)).reshape(N_c, 1), np.ones(N_c).reshape(N_c, 1), 2*np.ones(N_c).reshape(N_c, 1)))
K = 3 # Number of classes
N = K*N_c

# I_num = 3 # Number of input nodes + bias
# H_num = 3 # Number of hidden nodes

# w_t_ij = np.random.rand(I_num, H_num)
# w_t_jk = np.random.rand(H_num, K)
# y_j = np.zeros((3))
# y_k = np.zeros((K))
# delta_j = np.zeros((3))
# delta_k = np.zeros((K))
# ek = np.zeros((K))
# eta = 0.3 
# epochs = 1000


class NeuralNetwork_FFBP(object):
    def __init__(self, input_num, hidden_num, output_num, eta):
        self.input_num = input_num
        self.hidden_num = hidden_num
        self.output_num = output_num
        
        self.w_t_ij = np.random.rand(input_num, hidden_num)
        self.w_t_jk = np.random.rand(hidden_num, output_num)
        
        self.eta = eta

    def train(self, X, Y, epochs):
        
        for t in range(epochs):
            for i in range(X.shape[0]):
                x = X[i,:]

                target = np.zeros((self.output_num))
                target[int(Y[i])] = 1.0
                
                y, x_i, x_j = self._feed_forward(x)
                self._back_propagation(target, y, x_i, x_j)
                
        print "w_ij: {}".format(self.w_t_ij)
        print "w_jk: {}".format(self.w_t_jk)
        
    def predict(self,X):
        Y = []
        
        for i in range(X.shape[0]):
            x = X[i,:] 
            y, x_i, x_j = self._feed_forward(x)
            Y.append(np.argmax(y))
        
        return np.array(Y)
        
    def _feed_forward(self, x):
        y_j = np.zeros((self.hidden_num))
        y_k = np.zeros((self.output_num))
        
        x_i = x
        for j in range(self.hidden_num):
            A=np.dot(self.w_t_ij[:,j],x_i)
            y_j[j]=1.0/(1.0+np.exp(-A))

        x_j=y_j
        for k in range(self.output_num):
            A=np.dot(self.w_t_jk[:,k],x_j)
            y_k[k]=1.0/(1.0+np.exp(-A))
            
        return y_k, x_i, x_j
        
    def _back_propagation(self, target, y, x_i, x_j):        
        delta_j = np.zeros((self.hidden_num))
        delta_k = np.zeros((self.output_num))
        
        y_k = y
        for k in range(self.output_num):
            ek=target[k]-y_k[k]  
            delta_k[k] = (1.0-y_k[k])*y_k[k]*ek
                
        for j in range(self.hidden_num):
            delta_j[j] = (1.0-x_j[j])*x_j[j]*np.dot(self.w_t_jk[j,:],delta_k)
                
        for j in range(self.hidden_num):
            for k in range(self.output_num):
                self.w_t_jk[j,k] = self.w_t_jk[j,k]+self.eta*delta_k[k]*x_j[j]

        for i in range(self.input_num):
            for j in range(self.hidden_num):
                self.w_t_ij[i,j] = self.w_t_ij[i,j]+self.eta*delta_j[j]*x_i[i]
                
                

# Begin plotting here
# Define our class colors
cmap_light = ListedColormap(['#FFAAAA', '#AAAAFF', '#AAFFAA'])

# Generate the mesh
x_min, x_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
y_min, y_max = X[:, 2].min() - 0.5, X[:, 2].max() + 0.5
h = 0.02 # step size in the mesh
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
X_mesh = np.c_[np.ones((xx.size, 1)), xx.ravel(), yy.ravel()]
Z = np.zeros((xx.size, 1))

# Compute the likelihood of each cell in the mesh
# for i in range(0, xx.size):
#     for j in range(len(y_j)):
#         A=np.dot(w_t_ij[:,j], X_mesh[i, :])
#         y_j[j]=1.0/(1.0+np.exp(-A))

#     x_j=y_j
#     for k in range(len(y_k)):
#         A=np.dot(w_t_jk[:,k], x_j)
#         y_k[k]=1.0/(1.0+np.exp(-A))

#     Z[i] = np.argmax(y_k)

NN = NeuralNetwork_FFBP(3,3,3, 0.3)
NN.train(X,Y,200)
Z = NN.predict(X_mesh)

# Plot it
Z = Z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)
plt.plot(X[0:N_c-1, 1], X[0:N_c-1, 2], 'ro', X[N_c:2*N_c-1, 1], X[N_c:2*N_c-1,
    2], 'bo', X[2*N_c:, 1], X[2*N_c:, 2], 'go')
plt.axis([np.min(X[:, 1])-0.5, np.max(X[:, 1])+0.5, np.min(X[:, 2])-0.5, np.max(X[:, 2])+0.5])
plt.show()
```
