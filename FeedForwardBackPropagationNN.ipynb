```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# 1
x=np.array([1.0,3.0])
w=np.array([0.8,0.1])
bias=0.0
n=1.5
des=0.95

w_t=w
for i in range(100):
    A=np.dot(x,w_t)+bias
    y=1/(1+np.exp(-A))
    e=des-y
    w_t=w_t-n*(-e)*(1-y)*y*x

w_1 = w_t
x_1 = x
    
# 2
des = 0.05
w_t=w_1
x=np.array([-1.0,-3.0])
for i in range(10):
    A=np.dot(x,w_t)+bias
    y=1/(1+np.exp(-A))
    e=des-y
    w_t=w_t-n*(-e)*(1-y)*y*x
    
w_2 = w_t
x_2 = x
    
print 'Weights 2: {}'.format(w_2)

# Results [1,3]
A=np.dot(x_1,w_2)+bias
y_1=1/(1+np.exp(-A))

A=np.dot(x_2,w_2)+bias
y_2=1/(1+np.exp(-A))

print "y_1 = {}".format(y_1)
print "y_2 = {}".format(y_2)
```
