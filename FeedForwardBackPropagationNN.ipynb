```python
%matplotlib inline
import os, struct
from array import array as pyarray
from numpy import append, array, int8, uint8, zeros
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import cv2

# Load NIST (http://g.sweyla.com/blog/2012/mnist-numpy/)
def load_mnist(dataset="training", digits=np.arange(10), path="."):
    """
    Loads MNIST files into 3D numpy arrays

    Adapted from: http://abel.ee.ucla.edu/cvxopt/_downloads/mnist.py
    """

    if dataset == "training":
        fname_img = os.path.join(path, 'train-images-idx3-ubyte')
        fname_lbl = os.path.join(path, 'train-labels-idx1-ubyte')
    elif dataset == "testing":
        fname_img = os.path.join(path, 't10k-images-idx3-ubyte')
        fname_lbl = os.path.join(path, 't10k-labels-idx1-ubyte')
    else:
        raise ValueError("dataset must be 'testing' or 'training'")

    flbl = open('train-labels-idx1-ubyte\\train-labels.idx1-ubyte','rb')#fname_lbl, 'rb')
    magic_nr, size = struct.unpack(">II", flbl.read(8))
    lbl = pyarray("b", flbl.read())
    flbl.close()

    fimg = open('train-images-idx3-ubyte\\train-images.idx3-ubyte','rb')#fname_img, 'rb')
    magic_nr, size, rows, cols = struct.unpack(">IIII", fimg.read(16))
    img = pyarray("B", fimg.read())
    fimg.close()

    ind = [ k for k in range(size) if lbl[k] in digits ]
    N = len(ind)

    images = zeros((N, rows, cols), dtype=uint8)
    labels = zeros((N, 1), dtype=int8)
    for i in range(len(ind)):
        images[i] = array(img[ ind[i]*rows*cols : (ind[i]+1)*rows*cols ]).reshape((rows, cols))
        labels[i] = lbl[ind[i]]

    return images, labels


a = np.matrix([[1,3],[2,4]])
print np.matrix.sum(a)

images, labels = load_mnist('training', digits=[2])

#mean_img = images.mean(axis=0)
#mean_img = images[5].astype(np.float32)
mean_img = images[0]

print images.shape
print "Min: {}".format(mean_img.min())
print "Max: {}".format(mean_img.max())

print mean_img.dtype
#mean_img /= 255.0

div_img = abs(mean_img[0:14, 0:14]-mean_img[14:28, 14:28])
print "Sym: {}".format(np.sum(div_img))

ret,thresh = cv2.threshold(mean_img,127,255,0)
contours, hierarchy = cv2.findContours(thresh, 1, 2)
zero_img = np.zeros([28,28])
#cv2.drawContours(zero_img, contours[1], -1, (255,255,255), 1)
M = cv2.moments(contours[0])

cx = int(M['m10']/M['m00'])
cy = int(M['m01']/M['m00'])

print "cx={}".format(cx)
print "cy={}".format(cy)

# Convex hull defects
hull = cv2.convexHull(contours[1],returnPoints = False)
defects = cv2.convexityDefects(contours[1],hull)

for i in range(defects.shape[0]):
    s,e,f,d = defects[i,0]
    start = tuple(contours[1][s][0])
    end = tuple(contours[1][e][0])
    far = tuple(contours[1][f][0])
    #cv2.line(zero_img,start,end,[255,255,255],1)
    #cv2.circle(zero_img,far,1,[255,255,255],-1)
    
#plt.imshow(div_img, cmap=plt.cm.gray)
plt.imshow(mean_img, cmap=plt.cm.gray)
plt.Circle((cx,cy),2,color='r')
plt.show()

```

```python
# Generate three random clusters of 2D data
N_c = 200
A = 0.6*np.random.randn(N_c, 2)+[1, 1]
B = 0.6*np.random.randn(N_c, 2)+[3, 3]
C = 0.6*np.random.randn(N_c, 2)+[3, 0]
X = np.hstack((np.ones(3*N_c).reshape(3*N_c, 1), np.vstack((A, B, C))))
Y = np.vstack(((np.zeros(N_c)).reshape(N_c, 1), np.ones(N_c).reshape(N_c, 1), 2*np.ones(N_c).reshape(N_c, 1)))
K = 3 # Number of classes
N = K*N_c

# I_num = 3 # Number of input nodes + bias
# H_num = 3 # Number of hidden nodes

# w_t_ij = np.random.rand(I_num, H_num)
# w_t_jk = np.random.rand(H_num, K)
# y_j = np.zeros((3))
# y_k = np.zeros((K))
# delta_j = np.zeros((3))
# delta_k = np.zeros((K))
# ek = np.zeros((K))
# eta = 0.3 
# epochs = 1000


class NeuralNetwork_FFBP(object):
    def __init__(self, input_num, hidden_num, output_num, eta):
        self.input_num = input_num
        self.hidden_num = hidden_num
        self.output_num = output_num
        
        self.w_t_ij = np.random.rand(input_num, hidden_num)
        self.w_t_jk = np.random.rand(hidden_num, output_num)
        
        self.eta = eta

    def train(self, X, Y, epochs):
        
        for t in range(epochs):
            for i in range(X.shape[0]):
                x = X[i,:]

                target = np.zeros((self.output_num))
                target[int(Y[i])] = 1.0
                
                y, x_i, x_j = self._feed_forward(x)
                self._back_propagation(target, y, x_i, x_j)
                
        print "w_ij: {}".format(self.w_t_ij)
        print "w_jk: {}".format(self.w_t_jk)
        
    def predict(self,X):
        Y = []
        
        for i in range(X.shape[0]):
            x = X[i,:] 
            y, x_i, x_j = self._feed_forward(x)
            Y.append(np.argmax(y))
        
        return np.array(Y)
        
    def _feed_forward(self, x):
        y_j = np.zeros((self.hidden_num))
        y_k = np.zeros((self.output_num))
        
        x_i = x
        for j in range(self.hidden_num):
            A=np.dot(self.w_t_ij[:,j],x_i)
            y_j[j]=1.0/(1.0+np.exp(-A))

        x_j=y_j
        for k in range(self.output_num):
            A=np.dot(self.w_t_jk[:,k],x_j)
            y_k[k]=1.0/(1.0+np.exp(-A))
            
        return y_k, x_i, x_j
        
    def _back_propagation(self, target, y, x_i, x_j):        
        delta_j = np.zeros((self.hidden_num))
        delta_k = np.zeros((self.output_num))
        
        y_k = y
        for k in range(self.output_num):
            ek=target[k]-y_k[k]  
            delta_k[k] = (1.0-y_k[k])*y_k[k]*ek
                
        for j in range(self.hidden_num):
            delta_j[j] = (1.0-x_j[j])*x_j[j]*np.dot(self.w_t_jk[j,:],delta_k)
                
        for j in range(self.hidden_num):
            for k in range(self.output_num):
                self.w_t_jk[j,k] = self.w_t_jk[j,k]+self.eta*delta_k[k]*x_j[j]

        for i in range(self.input_num):
            for j in range(self.hidden_num):
                self.w_t_ij[i,j] = self.w_t_ij[i,j]+self.eta*delta_j[j]*x_i[i]
                
                

# Begin plotting here
# Define our class colors
cmap_light = ListedColormap(['#FFAAAA', '#AAAAFF', '#AAFFAA'])

# Generate the mesh
x_min, x_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
y_min, y_max = X[:, 2].min() - 0.5, X[:, 2].max() + 0.5
h = 0.02 # step size in the mesh
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
X_mesh = np.c_[np.ones((xx.size, 1)), xx.ravel(), yy.ravel()]
Z = np.zeros((xx.size, 1))

# Compute the likelihood of each cell in the mesh
# for i in range(0, xx.size):
#     for j in range(len(y_j)):
#         A=np.dot(w_t_ij[:,j], X_mesh[i, :])
#         y_j[j]=1.0/(1.0+np.exp(-A))

#     x_j=y_j
#     for k in range(len(y_k)):
#         A=np.dot(w_t_jk[:,k], x_j)
#         y_k[k]=1.0/(1.0+np.exp(-A))

#     Z[i] = np.argmax(y_k)

NN = NeuralNetwork_FFBP(3,3,3, 0.3)
NN.train(X,Y,200)
Z = NN.predict(X_mesh)

# Plot it
Z = Z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)
plt.plot(X[0:N_c-1, 1], X[0:N_c-1, 2], 'ro', X[N_c:2*N_c-1, 1], X[N_c:2*N_c-1,
    2], 'bo', X[2*N_c:, 1], X[2*N_c:, 2], 'go')
plt.axis([np.min(X[:, 1])-0.5, np.max(X[:, 1])+0.5, np.min(X[:, 2])-0.5, np.max(X[:, 2])+0.5])
plt.show()
```
